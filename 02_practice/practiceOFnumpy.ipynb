```python
# 18.
# there are total 3 technique to check the memory type (copy or view)
# a. base attribute
# b. flags object
# c. shares_memory() method 
anArr = np.arange(10)
anArr2 = anArr[::2]
print(anArr)
print(anArr2)
print(anArr2.base) 
# base attribute is use to check if the array is view or copy. if (array.base) value is none then it is copy
# else it will return the base array or the original full array
```

```python
# use of flags 
print(anArr.flags.owndata)      # return true/false   return true if have its own memory
print(anArr.flags.writeable)      # return true/false   return true if you can change the value          

```

```python
# use of shares_memory
np.shares_memory(anArr,anArr2)
```
