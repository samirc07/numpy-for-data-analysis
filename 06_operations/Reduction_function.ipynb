In NumPy, a **reduction function** is an operation that removes one or more dimensions from an array by performing a calculation along a specific axis.

While a standard Python reduction (like `functools.reduce`) typically results in a single scalar value, NumPy is much more powerful because it allows you to reduce across **rows**, **columns**, or the **entire multi-dimensional structure**.

---

## 1. How NumPy Reductions Work

Most NumPy reduction functions follow a common signature: `np.function(array, axis=None)`.

* **`axis=None` (Default):** Reduces the entire array into a single number.
* **`axis=0`:** Reduces down the **rows** (calculates a result for each column).
* **`axis=1`:** Reduces across the **columns** (calculates a result for each row).

### Code Example:

```python
import numpy as np

data = np.array([[1, 2, 3], 
                 [4, 5, 6]])

# Total reduction (everything)
print(np.sum(data))          # Output: 21

# Column-wise reduction (collapse rows)
print(np.sum(data, axis=0))  # Output: [5, 7, 9]

# Row-wise reduction (collapse columns)
print(np.sum(data, axis=1))  # Output: [6, 15]

```

---

## 2. Common Reduction Functions

NumPy provides a wide suite of these functions, all optimized in C for extreme speed.

| Function | Description |
| --- | --- |
| `np.sum()` | Adds all elements. |
| `np.prod()` | Multiplies all elements. |
| `np.mean()` | Calculates the arithmetic mean. |
| `np.std()` | Calculates the standard deviation. |
| `np.all()` | Logical AND (returns True if **all** elements are True). |
| `np.any()` | Logical OR (returns True if **any** element is True). |
| `np.argmin()` | Returns the **index** of the minimum value. |

# **6. Explain difference between sum() and Python sum().**
While they share the same name and goal, `numpy.sum()` and Python’s built-in `sum()` are worlds apart when it comes to how they handle data and memory.

---

## 1. Scope and Input

* **Python `sum()**`: Designed for **general-purpose iteration**. It works on any iterable (lists, tuples, dictionaries) and adds the elements one by one. It is optimized for standard Python objects.
* **NumPy `sum()**`: Specifically engineered for **numerical arrays**. It is aware of array dimensions (axes) and is written in C for massive performance gains.

---

## 2. Dimensionality (The "Axis" Power)

This is the biggest functional difference. Python's `sum()` has no concept of rows or columns; it just flattens everything it can.

* **Python `sum()**`: If you pass a 2D list to Python's `sum()`, it will likely throw an error because it tries to add a list to an integer.
* **NumPy `sum()**`: Can sum across specific directions using the `axis` parameter.

```python
import numpy as np

arr = np.array([[1, 2], [3, 4]])

# NumPy can sum columns
print(np.sum(arr, axis=0)) # Output: [4, 6]

# Python sum would fail on the 2D array directly
# sum(arr) -> TypeError

```

---

## 3. Performance (Speed)

Because NumPy uses **vectorization**, it performs the addition in compiled C code, bypassing the overhead of the Python interpreter for every single element.

| Data Size | Python `sum()` | NumPy `sum()` |
| --- | --- | --- |
| 1,000 items | Fast | Very Fast |
| 1,000,000 items | Slow (Seconds) | Instant (Milliseconds) |

> **The Golden Rule:** Never use Python's built-in `sum()` on a NumPy array. It is significantly slower because it has to "extract" each number from the NumPy format back into a Python object before adding it.

---

## 4. Handling Missing Data

* **Python `sum()**`: If there is a `None` or a `NaN` (Not a Number), it will usually crash.
* **NumPy `sum()**`: If there is a `NaN`, the result will be `NaN`. However, NumPy also provides `np.nansum()`, which ignores missing values and sums the rest.

---

### Summary Comparison Table

| Feature | Python `sum(iterable)` | `numpy.sum(array)` |
| --- | --- | --- |
| **Data Type** | Lists, Tuples, Sets | NumPy Arrays |
| **Algorithm** | Sequential (Looping) | Vectorized (Parallel-like) |
| **Multi-dimensional** | No | Yes (via `axis`) |
| **Speed** | Standard | Blazing Fast |

**13.Why does std() return float even for int arrays?**

Standard deviation returns a float because the division (for the mean) and the square root (for the final value) rarely result in whole numbers, and using integers would cause significant loss of precision.


````python
14.
Arr = np.array([34,52,76,52,5,72,67,85,68,79,89])

first_mean = np.mean(Arr)
second_mean = Arr.mean()

print(f"first mean is : {first_mean}")
print(f"second mean is : {second_mean}")
````
