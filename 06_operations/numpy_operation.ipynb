| Feature           | `np.dot(a, b)`        | `a * b`                       |
| ----------------- | --------------------- | ----------------------------- |
| Type of operation | Linear algebra        | Element-wise                  |
| Shape rules       | Matrix rules apply    | Same shape (or broadcastable) |
| Result meaning    | Sum of products       | Direct multiplication         |
| Used in           | ML, DL, math, physics | Data scaling, masking         |

Element-wise means:
👉 Operate on each position independently.

dot product
a = [1, 2, 3]
b = [4, 5, 6]

a · b = (1×4) + (2×5) + (3×6) = 32




element wise 
a = [1, 2, 3]
b = [4, 5, 6]

a * b   →   [1*4, 2*5, 3*6]

```python
d = np.array([1,2,3,4])
e = np.array([1,2,3,4])
np.array_equal(d,e)
# array_equal method compare the each index, if every element is same then return true
# and if any one element is not same then return false 
```

```python
# sum method give sum of all element 
print(d.sum())

ar = np.array([[1,2,3],[4,5,6]])
print(ar.sum(axis=0))  # sum of columns 
print(ar.sum(axis=1))   # sum of rows 
```

```python
# other methods
print("for minimum element in array :",ar.min())
print("for maximum element in array :",ar.max())
print()
print("index of minimum element : ", ar.argmin())
print("index of maximum element : ", ar.argmax())

```

```python
# logical operation 

print(np.all([True,True,False])) # if all are true then true 
print(np.any([True,True,False])) # if all are false then false 
```

```python
# use of all method 
a = np.array([1,2,5,2,6])
b = np.array([9,74,33,8,22])
c = np.array([100,200,300,400,500])

((a < b) & (b < c)).all()  # mean if condition is true for all then its true else false 
```

**Broadcasting**

basic operation on numpy arrays are elemeentwise (possible if size of arrays are same)

its possible to do operations of arrays with different shape or size if numpy can transform these arrays
into same size, this method is called broadcasting

Example:
A (3×3 array)        B (1×3 array)

A =                 B =
┌───┬───┬───┐       ┌───┬───┬───┐
│ 1 │ 2 │ 3 │       │10 │20 │30 │
├───┼───┼───┤
│ 4 │ 5 │ 6 │
├───┼───┼───┤
│ 7 │ 8 │ 9 │
└───┴───┴───┘


Broadcasting Process
────────────────────

B is stretched (NOT copied in memory):

┌───┬───┬───┐
│10 │20 │30 │
├───┼───┼───┤
│10 │20 │30 │
├───┼───┼───┤
│10 │20 │30 │
└───┴───┴───┘


Element-wise Operation (A + B)
─────────────────────────────

┌───┬───┬───┐
│11 │22 │33 │
├───┼───┼───┤
│14 │25 │36 │
├───┼───┼───┤
│17 │28 │39 │
└───┴───┴───┘


```python
# tile method is use for broadcasting 
arry = np.array([1,2,3,4])
arry = np.tile(arry,(5,1))  # (strech row , strech column)
print(arry)
```

```python
# we can transpose the matrix (row become column and column become row)
arry.T
```

```python
arry2 = np.array([4,5,6,7])
```

```python
arry+arry2   # broadcasting happen internally and broadcasting happen according to column
```

```python
# first = np.array([[1,2,4],[]])   we can't create a single column array like this 
first = np.arange(0,50,10)
first = first[:,np.newaxis]  # creating 2d array 
print(first)
print(first.shape)
```

```python
# reshape method 
for_reshape = np.array([1,2,3,4,5,6])
for_reshape = for_reshape.reshape(6,1)
print("converting simple array into a matrix using reshapce method")
print(for_reshape)

# changing again into a normal array using ravel method 
for_reshape = for_reshape.ravel()
print("Converting matrix into simple array using ravel method")
print(for_reshape)
```
