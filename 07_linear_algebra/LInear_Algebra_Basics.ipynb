The dot product is a fundamental operation between two vectors that yields a scalar value, capturing both their magnitudes and the angle between them.

## Geometric Definition
The dot product of vectors \(\vec{a}\) and \(\vec{b}\) is \(\vec{a} \cdot \vec{b} = |\vec{a}| \, |\vec{b}| \cos \theta\), where \(\theta\) is the angle between them  [cuemath](https://www.cuemath.com/algebra/dot-product/). This formula shows it equals zero when vectors are perpendicular (\(\cos 90^\circ = 0\))  [cuemath](https://www.cuemath.com/algebra/dot-product/). For parallel vectors, it reaches a maximum equal to the product of their lengths  [cuemath](https://www.cuemath.com/algebra/dot-product/).

## Algebraic Formula
For vectors \(\vec{a} = (a_1, a_2, \dots, a_n)\) and \(\vec{b} = (b_1, b_2, \dots, b_n)\), the dot product is \(\vec{a} \cdot \vec{b} = a_1 b_1 + a_2 b_2 + \dots + a_n b_n\). In 3D with unit vectors, \(\vec{a} = a_1 \hat{i} + a_2 \hat{j} + a_3 \hat{k}\) and \(\vec{b} = b_1 \hat{i} + b_2 \hat{j} + b_3 \hat{k}\) gives \((a_1 b_1) + (a_2 b_2) + (a_3 b_3)\) since orthogonal unit vectors dot to zero. [cuemath](https://www.cuemath.com/algebra/dot-product/)

## Key Properties
- **Commutative**: \(\vec{a} \cdot \vec{b} = \vec{b} \cdot \vec{a}\). [cuemath](https://www.cuemath.com/algebra/dot-product/)
- **Distributive**: \(\vec{a} \cdot (\vec{b} + \vec{c}) = \vec{a} \cdot \vec{b} + \vec{a} \cdot \vec{c}\). [cuemath](https://www.cuemath.com/algebra/dot-product/)
- It projects one vector onto another, useful in physics for work calculations. For example, \(\vec{a} = (2, 3, 1)\), \(\vec{b} = (5, -2, 3)\) yields \(2\cdot5 + 3\cdot(-2) + 1\cdot3 = 7\). [math.libretexts](https://math.libretexts.org/Bookshelves/Calculus/Calculus_(OpenStax)/12:_Vectors_in_Space/12.03:_The_Dot_Product)

In NumPy, `dot()` and the `@` operator both compute matrix products but differ in behavior and intent, especially across array dimensions.

## Core Functionality
`numpy.dot(a, b)` handles dot products flexibly: for 1D arrays, it's a scalar dot product (e.g., sum of element-wise products); for 2D, matrix multiplication; for higher dimensions, it sums products over the last axis of `a` and second-to-last of `b`. [blog.finxter](https://blog.finxter.com/numpy-operator-vs-np-dot/)
The `@` operator (Python 3.5+) calls `numpy.matmul(a, b)`, designed strictly for matrix multiplication, treating 1D arrays as row/column vectors and enabling clear batch operations on higher-D arrays. [stackoverflow](https://stackoverflow.com/questions/34142485/difference-between-numpy-dot-and-python-3-5-matrix-multiplication)
Both yield identical results for 2D arrays, like `np.array([[1,2],[3,4]]) @ np.array([[5,6],[7,8]])` equals `np.dot(...)`. [blog.finxter](https://blog.finxter.com/numpy-operator-vs-np-dot/)

## Key Differences
| Aspect | `np.dot()` | `@` (or `np.matmul()`) |
|--------|------------|-------------------------|
| 1D Arrays | Scalar (e.g., `[1,2] @ [3,4]` → 11)  [stackoverflow](https://stackoverflow.com/questions/34142485/difference-between-numpy-dot-and-python-3-5-matrix-multiplication/34142617) | Scalar, same as dot  [community.deeplearning](https://community.deeplearning.ai/t/np-matmul-vs-np-dot/217450) |
| 2D Arrays | Matrix mult. | Matrix mult., more readable  [blog.finxter](https://blog.finxter.com/numpy-operator-vs-np-dot/) |
| 3D+ Arrays | Sums last axis of first with second-last of second  [stackoverflow](https://stackoverflow.com/questions/34142485/difference-between-numpy-dot-and-python-3-5-matrix-multiplication/34142617) | Batch matrix mult. (e.g., stacks matrices)  [codecademy](https://www.codecademy.com/article/numpy-matrix-multiplication-a-beginners-guide) |
| Readability | Verbose: `np.dot(a, b)` | Clean: `a @ b`  [blog.finxter](https://blog.finxter.com/numpy-operator-vs-np-dot/) |

## When to Use
Prefer `@` for matrix math in modern code—it's explicit and chainable (e.g., `A @ B @ C`). [blog.finxter](https://blog.finxter.com/numpy-operator-vs-np-dot/)
Use `dot()` for legacy code (pre-Python 3.5) or when needing its unique higher-D summation. [stackoverflow](https://stackoverflow.com/questions/34142485/difference-between-numpy-dot-and-python-3-5-matrix-multiplication/34142617)
Example: Vectors `u = np.array([1,2])`, `v = np.array([3,4])`; both give `11`, linking to geometric dot product \(1\cdot3 + 2\cdot4\). [blog.finxter](https://blog.finxter.com/numpy-operator-vs-np-dot/)
